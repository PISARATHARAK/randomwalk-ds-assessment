# Q1: How many rows and columns are there in books.csv dataset?

```python
rows,col=df.shape
print("In books.cvs dataset there are ",rows,"rows and ",col,"cols");
```

# Q2:How many books do not have an original title?

```python
#Method1 using inbuilt function
null_values = df['original_title'].isnull().sum()
print(null_values,"books do not have an original title")

#Method2 using loops
null_count = 0

# Loop through each row in the DataFrame
for i in df.index:
    if pd.isnull(df['original_title'][i]):
        null_count += 1
print("Using for loops:")
print(null_count,"books do not have an original title")
```

# Q3: How many unique books are present in the dataset ? Evaluate based on the 'book_id' after removing null values in the original_title column.

```python
df1=df.dropna(subset=['original_title'])
#Checking whether there are null values in original title
null_values = df1['original_title'].isnull().sum()
print(null_values,"books do not have an original title")

unique_ids = df1['book_id'].unique()  #this will return unique book id list 

no_of_unique_books=len(unique_ids)
print("unique books present in the dataset are:",no_of_unique_books)
```

# Q4: What is the average rating of all the books in the dataset based on ‘average_rating’?

```python
average_rating = df['average_rating'].mean()
print("The average rating of all the books in the dataset based on ‘average_rating’ is:",average_rating)

#Using loops
Sum=0
count=0
for i in df.index:
    Sum+=df['average_rating'][i]
    count+=1
avg_rating=Sum/count
print("The average rating of all the books in the dataset based on ‘average_rating’ is:",avg_rating)
```

# Q5: Find the number of books published in the year ‘2000’ based on the ‘original_publication_year’.

```python
books_2000=df.loc[df['original_publication_year']==2000]
rows,cols=books_2000.shape
print("The number of books published in the year ‘2000’ based on the ‘original_publication_year’ are:",rows)

#Using loops
no_books=0
for i in df.index:
    if df['original_publication_year'][i]==2000:
        no_books+=1;
print("The number of books published in the year ‘2000’ based on the ‘original_publication_year’ are:",no_books)
```

# Q6: Which book (title) has the maximum number of ratings based on ‘work_ratings_count’.

```python
max_ratings=df['work_ratings_count'].max()

book_with_max_ratings = df.loc[df['work_ratings_count'] == max_ratings, 'title'].iloc[0]
print("The book with the maximum number of ratings is:", book_with_max_ratings)
```

# Q7: Bucket the average_rating of books into 11 buckets [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0] with 0.5 decimal rounding (eg: average_rating 3.0 to 3.49 will fall in bucket 3.0). Plot bar graph to show total number of books in each rating bucket.

```python
import matplotlib.pyplot as plt
bins=[0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0]
plt.hist(df['average_rating'], bins=bins, edgecolor='black')
plt.xlabel('Average Rating')
plt.ylabel('Number of Books')
plt.title('Histogram of Average Ratings')
plt.xticks(bins)

plt.show()
```
